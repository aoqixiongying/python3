### pandas索引

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df['Age'][:5]
```

```python
df[['Age', 'Fare']][:5]
```

* loc 用label来去定位
* iloc 用position进行定位

```python
df.iloc[0:5, 1:3]
```

```python
df = df.set_index('Name')
df.loc['Heikkinen, Miss. Laina']
```

```python
df.loc['Heikkinen, Miss. Laina', 'Fare']
```

```python
df.loc['Heikkinen, Miss. Laina':'Allen, Mr. William Henry',:]
```

```python
df.loc['Heikkinen, Miss. Laina', 'Fare'] = 1000
df.head()
```

```python
# deprecationWarning
df.ix['Heikkinen, Miss. Laina', 'Fare']
```

bool类型的索引

```python
df['Fare'] > 40
```

```python
df[df['Fare'] > 40]
```

```python
df[df['Sex'] == 'male'][:5]
```

```python
df.loc[df['Sex'] =='male', 'Age'].mean()
```
